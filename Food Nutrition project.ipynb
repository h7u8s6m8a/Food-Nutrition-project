## Feature Engineering

### Sugar Ratio Analysis

```python
df["sugar_ratio"] = df["freesugar_g"] / df["carb_g"] *100
df["sugar_ratio"]
```

### Protein Density Analysis

```python
df["protein_density"] = df["protein_g"] / df["energy_kcal"] *100
df["protein_density"]
```

### Healthy Score Analysis

```python
df["healthy_score"] = (df["protein_g"] +df["fibre_g"] - df["freesugar_g"] -df["fat_g"])
df["healthy_score"]
```

### TOP 10 HEALTHY FOODS

```python
df.sort_values("healthy_score", ascending=False)[["food_name","healthy_score"]].head(10)
```

### Food Category

```python
def label_food(row):
    if row["protein_g"] > 15 and row["fat_g"] < 10:
        return "Muscle Gain"
    elif row["energy_kcal"] < 200:
        return "Weight Loss"
    else:
        return "General Health"

df["food_category"] = df.apply(label_food, axis=1)
df["food_category"]
```

### Create Risk Flags

```python
df["high_sugar_flag"] = df["freesugar_g"] > 10
df["high_cholesterol_flag"] = df["cholesterol_mg"] > 150
df["high_sfa_flag"] = df["sfa_mg"] > 5000
df["low_fiber_flag"] = df["fibre_g"] < 2
```

### Create Positive Nutrition Flags

```python
df["good_protein_flag"] = df["protein_density"] >= 0.1
df["godd_fiber_flag"] = df["fibre_g"] >= 5
```

```python
def categorize_food(row):
    if(
        row["high_sugar_flag"] or 
        row["high_cholesterol_flag"] or
        row["high_sfa_flag"]
    ):
        return "Unhealthy"
    elif(
        row["good_protein_flag"] and
        row["godd_fiber_flag"]
    ):
        return "Healthy"
    else:
        return "Moderate"
```

```python
df["food_category-2"] = df.apply(categorize_food, axis=1)
df["food_category-2"]
```
